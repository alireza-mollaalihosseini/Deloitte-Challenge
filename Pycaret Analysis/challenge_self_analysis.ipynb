```python
diesel_flags = ["Q8","Api-Ip","Total Erg","Enercoop","Repsol","Vega","Agip Eni"]
diesel_provinces = ["MI","PA","LI","NP","SO","PD","VE","CE","PR","BG"]

benzina_flags = ["Agip Eni","Acton1Q","Tamoil","San Marco Petroli","Vega","Sarni Oil","Pompe Bianche","Enercoop"]
benzina_provinces = ["BZ","CO","TN","CT","BO","RM","IM","VA","RG","VB"]

metano_flags = ["Tamoil","Agip Eni","Esso","KEROPETROL","Pompe Bianche","Vega","Simonetti Petroli"]
metano_provinces = ["RA","FG","CN","CS","VA","TR","EN","BZ","VE","IM"]

gpl_flags = ["San Marco Petroli","GasAutoSud","Sia fuel","Pompe Bianche","Blanco petroli","Api-Ip","Q8","Retitalia"]
gpl_provinces = ["NO","AN","RI","IM","GE","PZ","LE","CE","BT","VI"]
# The Intensity of Importance goes to 60(while in other categories it goes to 10 at most)

gasolio_flags = ["Tamoil","Esso","Sarni Oil","Api-Ip","Agip Eni","Pompe Bianche"]
gasolio_provinces = ["BZ","BL","NP","TN","GE","CO","AO","IM","SA","MB"]

selected_diesel  = self_data[(self_data.carburante=="diesel") & ((self_data.Bandiera.isin(diesel_flags)) | (self_data.Provincia.isin(diesel_provinces)))][["prezzo","Bandiera","Provincia","Latitudine","Longitudine"]]
selected_benzina = self_data[(self_data.carburante=="benzina") & ((self_data.Bandiera.isin(benzina_flags)) | (self_data.Provincia.isin(benzina_provinces)))][["prezzo","Bandiera","Provincia","Latitudine","Longitudine"]]
selected_metano  = self_data[(self_data.carburante=="metano") & ((self_data.Bandiera.isin(metano_flags)) | (self_data.Provincia.isin(metano_provinces)))][["prezzo","Bandiera","Provincia","Latitudine","Longitudine"]]
selected_gpl     = self_data[(self_data.carburante=="gpl") & ((self_data.Bandiera.isin(gpl_flags)) | (self_data.Provincia.isin(gpl_provinces)))][["prezzo","Bandiera","Provincia","Latitudine","Longitudine"]]
selected_gasolio = self_data[(self_data.carburante=="gasolio") & ((self_data.Bandiera.isin(gasolio_flags)) | (self_data.Provincia.isin(gasolio_provinces)))][["prezzo","Bandiera","Provincia","Latitudine","Longitudine"]]
```
